```python
def replace_marital(val):
    if val =='single':
        return 0
    else:
        return 1
data['marital'] = data['marital'].apply(replace_marital,1)
```

```python
data.head()
```

```python
def replace_housing(hus):
    if hus =='no':
        return 0
    else:
        return 1
data['housing'] = data['housing'].apply(replace_housing,1)
```

```python
data.head()
```

```python
data['loan'] = data['loan'].replace({
    'no':0,
    'yes':1
})
```
