```python
bins=[0,5,10,15,20,25,30,35,40,45,50,55]
plt.figure(figsize=(10,8))
sns.set()
sns.distplot(df['total_bill'], bins=bins)
plt.xticks(bins)
plt.title("Histogram of total Bill")
```

```python
plt.figure(figsize=(10,8))
sns.set()
sns.distplot(df['total_bill'], hist_kws ={'color':'#666666','edgecolor':'#555888',
                                         'Linewidth':2, 'linestyle':'--','alpha':.9})
```

```python
plt.figure(figsize=(10,8))
sns.set()
sns.distplot(df['total_bill'], hist_kws ={'color':'#666666','edgecolor':'#555888',
                                         'Linewidth':2, 'linestyle':'--','alpha':.9},
            kde_kws={'color':'#666666','Linewidth':2, 'linestyle':'--','alpha':.9})
```

```python
plt.figure(figsize=(10,8))
sns.set()
sns.distplot(df['total_bill'], hist_kws ={'color':'#888888','edgecolor':'#555888',
                                         'Linewidth':2, 'linestyle':'--','alpha':.9},
            kde_kws={'color':'#666666','Linewidth':4, 'linestyle':'--','alpha':.9},
            rug =True,
            rug_kws = {'color':'#888888','edgecolor':'#555888',
                                         'Linewidth':2, 'linestyle':'--','alpha':.9})
```

```python
plt.figure(figsize=(10,8))
sns.set()
sns.distplot(df['total_bill'], hist_kws ={'color':'#888888','edgecolor':'#555888',
                                         'Linewidth':2, 'linestyle':'--','alpha':.9},
            kde = False,
            fit = norm,
            fit_kws = {'color':'#222222','Linewidth':2, 'linestyle':'--','alpha':.9},
            rug =True,
            rug_kws = {'color':'#888888','edgecolor':'#111111',
                                         'Linewidth':2, 'linestyle':'--','alpha':.9})
```

```python
sns.set()
plt.figure(figsize=(10,8))
sns.distplot(df['total_bill'], bins=9, label ='Total Bill')
sns.distplot(df['tip'], bins =9, label='Tip')
sns.distplot(df['size'], bins =9, label='Size')
plt.legend()
```
