```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
sns.barplot(x='time',y='total_bill', data=df, hue='sex')
```

```python
sns.lineplot(x='total_bill',y='tip',data=df)
```

```python
sns.lineplot(x='tip',y='total_bill',data=df)
```

```python
sns.lineplot(x='tip',y='size',data=df)
```

```python
sns.lineplot(x='size',y='total_bill',data=df)
```

```python
sns.lineplot(x='size', y='total_bill', data=df, hue='sex')
plt.legend(loc='upper left')
```

```python
sns.lineplot(x='size', y='total_bill', data=df, hue='sex', style='sex', palette='Accent')
plt.legend(loc='upper left')
plt.show()
```

```python
sns.lineplot(x='size', y='total_bill', data=df, hue='sex', style='sex', palette='hot', 
             dashes=False, markers=['o','<'])
plt.legend(loc='upper left')
plt.show()
```

```python
sns.set(style='darkgrid')
sns.lineplot(x='size', y='total_bill', data=df, hue='sex', style='sex', palette='hot', 
             dashes=False, markers=['o','<'])

plt.title("Line Plot")
plt.xlabel('Size',fontsize=15)
plt.ylabel('Total Bill', fontsize=15)
plt.legend(loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.set(style='darkgrid')
sns.lineplot(x='size', y='total_bill', data=df, hue='sex', style='sex', palette='hot', 
             dashes=False, markers=['o','<'])

plt.title("Line Plot")
plt.xlabel('Size',fontsize=15)
plt.ylabel('Total Bill', fontsize=15)
plt.legend(loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.set(style='darkgrid')
sns.lineplot(x='size', y='total_bill', data=df, hue='day', style='day', palette='hot', 
             dashes=False, markers=['o','<','>','^'])

plt.title("Line Plot")
plt.xlabel('Size',fontsize=15)
plt.ylabel('Total Bill', fontsize=15)
plt.legend(loc='upper left')
plt.show()
```
