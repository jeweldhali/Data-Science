```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
sns.boxplot(x='size', data=df)
```

```python
sns.boxplot(x='total_bill', data=df, notch =True)
```

```python
sns.set()
```

```python
sns.boxplot(x='total_bill', data=df)
```

```python
sns.boxplot(x='sex',y='total_bill', data=df)
```

```python
sns.boxplot(x='day',y='total_bill', data=df)
```

```python
sns.boxplot(x='day',y='total_bill', data=df, hue ='sex')
```

```python
sns.boxplot(x='day',y='total_bill', data=df, hue ='sex', palette='husl')
```

```python
sns.boxplot(x='day',y='total_bill', data=df, hue ='smoker', palette='coolwarm')
```

```python
sns.boxplot(x='day',y='total_bill', data=df, hue ='time')
plt.legend(loc='upper left')
```

```python
sns.boxplot(x='day',y='total_bill', data=df, order=['Sun','Fri','Thur','Sat'])
plt.legend(loc='upper left')
```

```python
sns.boxplot(x='sex',y='tip', data=df)
```
