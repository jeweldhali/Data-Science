```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = sns.load_dataset('flights')
df.head()
```

```python
df.shape
```

```python
sns.set()
sns.barplot(x='year',y='passengers', data=df)
```

```python
df1 = df.pivot('month','year','passengers')
df1
```

```python
sns.heatmap(df1)
```

```python
sns.heatmap(df1, annot=True)
```

```python
sns.heatmap(df1, annot=True,fmt='d')
```

```python
sns.heatmap(df1, annot=True, fmt='d', linewidths=.9)
```

```python
sns.heatmap(df1, annot=True, fmt='d', linewidths=.9, vmin=100, vmax=650)
```

```python
sns.heatmap(df1, annot=True, fmt='d', linewidths=.9, cmap='summer')
```

```python
sns.heatmap(df1, annot=True, fmt='d', linewidths=.9, cmap='winter_r')
```

```python
sns.heatmap(df1, center=df1.loc['June',1956], annot=True, fmt='d', linewidths=.9)
```

```python
sns.heatmap(df1, center=df1.loc['March',1959], annot=True, fmt='d', linewidths=.9)
```

```python
sns.heatmap(df1, center=df1.loc['March',1959], annot=True, fmt='d', linewidths=.9, cbar=False)
```

```python
cor = df1.corr()
cor
```

```python
sns.heatmap(cor)
```
