```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = sns.load_dataset('tips')
```

```python
data.head()
```

```python
sns.scatterplot(x='total_bill', y='tip', data=data)
plt.show()
```

```python
sns.regplot(x='total_bill', y='tip', data=data)
plt.show()
```

```python
sns.regplot(x='total_bill', y='tip', data=data, color='red')
plt.show()
```

```python
sns.barplot(x='total_bill', y='tip', data=data)
plt.show()
```

```python
data.max()
```

```python
mean = [2,3]
cov = [[1.1,.4],[2.2,3]]

x_value, y_value = np.random.multivariate_normal(mean,cov,100).T

sns.regplot(x=x_value,y=y_value)
plt.show()
```

```python
sns.regplot(x=x_value,y=y_value, color='purple')
plt.show()
```

```python
series_x = pd.Series(x_value, name = 'Var-x')
series_y = pd.Series(y_value, name = 'Vay-y')
sns.regplot(x=series_x, y=series_y)
```

```python
sns.regplot(x=series_x, y=series_y, marker='D')
```

```python
sns.regplot(x=series_x, y=series_y, marker='D', color='gray')
```

```python
sns.regplot(x='total_bill', y='tip', data=data, marker='+',
           scatter_kws={'color':'blue'},
           line_kws={'color':'red','linewidth':3.1})
plt.show()
```

```python
sns.regplot(x='total_bill',y='tip',data=data,ci=14)
```

```python
sns.regplot(x='total_bill',y='tip',data=data,ci=100)
```

```python
sns.regplot(x='size',y='total_bill',data=data)
```

```python
sns.regplot(x='size',y='total_bill',data=data,x_jitter=.13, scatter_kws={'color':'red'})
```

```python
sns.regplot(x='size',y='total_bill',data=data, x_estimator=np.mean)
```
