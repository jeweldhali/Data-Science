```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pounds = [10,20,12,11,15]
snack = ['Cake','Nuts','Chips','Popcorn','Pretzels']
```

```python
plt.pie(pounds)
plt.show()
```

```python
plt.pie(pounds,labels=snack)
plt.show()
```

```python
plt.axis('equal')
plt.pie(pounds,labels=snack)
plt.show()
```

```python
plt.axis('equal')
plt.pie(pounds,labels=snack, radius=1.5)
plt.show()
```

```python
plt.axis('equal')
plt.pie(pounds,labels=snack, radius=1.5, autopct='%.0f%%')
plt.show()
```

```python
plt.axis('equal')
plt.pie(pounds,labels=snack, radius=1.5, autopct='%.0f%%',shadow=True)
plt.show()
```

```python
plt.axis('equal')
plt.pie(pounds,labels=snack, radius=1.5, autopct='%.0f%%',shadow=True,
       explode=[0,0,0,0,0])
plt.show()
```

```python
plt.axis('equal')
plt.pie(pounds,labels=snack, radius=1.5, autopct='%.0f%%',shadow=True,
       explode=[0,.3,0,0,0])
plt.show()
```

```python
plt.axis('equal')
plt.pie(pounds,labels=snack, radius=1.5, autopct='%.0f%%',shadow=True,
       explode=[0,.3,0,0,.2])
plt.show()
```

```python
plt.axis('equal')
plt.pie(pounds,labels=snack, radius=1.5, autopct='%.0f%%',shadow=True,
       explode=[0,0,0,0,0], startangle=160)
plt.show()
```

```python
color = ['r','g','m','b','c']
```

```python
plt.axis('equal')
plt.pie(pounds,labels=snack, radius=1.5, autopct='%.0f%%',shadow=True,
       explode=[0,0,0,0,0], colors = color)
plt.legend(loc=('28,5'))
plt.show()
```
