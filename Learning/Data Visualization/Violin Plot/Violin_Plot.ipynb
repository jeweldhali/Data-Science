```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
sns.violinplot(x='tip',data=df)
```

```python
sns.violinplot(x='total_bill',data=df)
```

```python
sns.violinplot(x='size', y = 'total_bill',data=df)
```

```python
sns.violinplot(x='day', y = 'total_bill',data=df)
```

```python
sns.violinplot(x='day', y = 'total_bill',data=df, hue='sex')
```

```python
sns.violinplot(x='day', y = 'total_bill',data=df, hue='sex', palette='coolwarm')
```

```python
sns.violinplot(x='day', y = 'total_bill',data=df, hue='smoker')
```

```python
sns.violinplot(x='day', y = 'total_bill',data=df, hue='smoker', split=True)
```

```python
sns.violinplot(x='day', y = 'total_bill',data=df, hue='smoker', split=True,
              order =['Sat','Thur','Fri','Sun'])
```

```python
sns.violinplot(x= 'day', y='total_bill', data=df,inner='quartile', hue='smoker')
```

```python
sns.violinplot(x= 'day', y='total_bill', data=df,inner='quartile', hue='smoker', split=True)
```

```python
sns.violinplot(x= 'day', y='total_bill', data=df,inner='quartile', hue='sex', split=True)
```

```python
sns.violinplot(x= 'day', y='total_bill', data=df,inner='quartile', hue='sex', scale='count')
```

```python
sns.violinplot(x= 'day', y='total_bill', data=df,inner='stick', hue='sex')
```

```python
sns.violinplot(x= 'day', y='total_bill', data=df,inner='quartile', hue='sex', split=True, bw=.1)
```
