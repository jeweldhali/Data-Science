```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
df2 = sns.load_dataset('iris')
```

```python
df2.head()
```

```python
sns.jointplot(x='total_bill',y='tip',data=df)
```

```python
sns.scatterplot(x='total_bill',y='tip',data=df)
```

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=df2)
```

```python
sns.set()
sns.jointplot(x='total_bill',y='tip',data=df, kind='reg')
```

```python
sns.set()
sns.jointplot(x='total_bill',y='tip',data=df, kind='reg', color='green')
```

```python
sns.set()
sns.jointplot(x='total_bill',y='tip',data=df, kind='hex')
```

```python
sns.set()
sns.jointplot(x='total_bill',y='tip',data=df, kind='kde')
```

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=df2, kind='kde')
```

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=df2, kind='kde', color='red')
```

```python
from scipy.stats import spearmanr
```

```python
sns.jointplot(x='total_bill',y='tip',data=df, stat_func =spearmanr)
```

```python
sns.jointplot(x='size',y='total_bill',data=df, stat_func =spearmanr, color='blue')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df, ratio=4, size=6)
```

```python
sns.jointplot(x='total_bill',y='tip',data=df).plot_joint(sns.kdeplot,n_levels=6)
```

```python
(sns.jointplot(x='total_bill',y='tip',data=df)
 .plot_joint(sns.kdeplot,n_levels=6)
 .plot_marginals(sns.rugplot))
```
